```python
import glob
```

```python
files_wo_extension = []
for file in glob.glob('*/*.ipynb'):
    file_wo_extension = file.split('.')[0]
    files_wo_extension.append(file_wo_extension)
```

```python
import itertools as itr
```

```python
files_grouped = itr.groupby(sorted(files_wo_extension), key=lambda file: file.split('/')[0])
```

```python
for folder, group in files_grouped:
    for idx, file in enumerate(group):
        if idx == 0:
            print(f'- file: {file}')
            print('  sections:')
        print(f'  - file: {file}')
    print()
```
